## Read in list of zip codes

```python
import pandas as pd
```

```python
zip_codes = pd.read_csv('/Users/myradeng/Documents/20220605 NYC Zips.csv',header=None)
```

```python
list_of_zip_codes = list(zip_codes[0])
```

```python
list_of_zip_codes
```

## Request data from census API call

```python
import requests
#api_url = "http://api.census.gov/data/2019/cbp?get=EMP&for=county:*&in=state:36"
bad_zip_codes = []
df_list = []
for zip_code in set(list_of_zip_codes):
    #api_url = "http://api.census.gov/data/2019/cbp?get=EMP&for=county:*&in=state:36"
    api_url = "http://api.census.gov/data/2020/cbp?get=EMP,EMP_N,NAICS2017,GEO_ID,GEOCOMP,ESTAB,EMPSZES,INDGROUP,SECTOR,SUBSECTOR&for=zipcode:" + str(zip_code)
    #api_url = "http://api.census.gov/data/2020/cbp?get=EMP,EMP_N,NAICS2017,PAYANN,PAYANN_N,ucgid,GEO_ID,GEOCOMP,ESTAB,EMPSZES,INDGROUP,SECTOR,SUBSECTOR&for=zipcode:" + str(zip_code)
    #api_url = "http://api.census.gov/data/2020/cbp?get=ESTAB,LFO,NAICS2017_LABEL,NAME&for=state:06&NAICS2017=72"
    response = requests.get(api_url)
    try: 
        df = pd.DataFrame(response.json())
        df.columns = df.iloc[0]
        df = df.drop([0])
        df_list.append(df)
        print("Processing zip code: " + str(zip_code))
    except ValueError:
        bad_zip_codes.append(zip_code)
        print("No data for zipcode: " + str(zip_code))
```

```python
total_df = pd.concat(df_list)
```

```python
total_df['EMP'] = total_df['EMP'].astype(int)
```

```python
total_df.to_csv('/Users/myradeng/Documents/census_business_data_2.csv')
```
